```python
%matplotlib inline
import pandas as pd
```

```python
titles = pd.read_csv('data/titles.csv')  ##### Will be available on Github  - https://github.com/DataScience75/Top_mentor_projects_Datasets/blob/main/titles.csv
titles.head()
```

```python
cast = pd.read_csv('data/cast.csv')
cast.head()
```

### What are the ten most common movie names of all time?

```python
titles['title'].value_counts().head(10)
```

### Which three years of the 1930s saw the most films released?

```python
movies_1930s = titles[(titles['year'] >= 1930) & (titles['year'] <= 1939)]
films_per_year = movies_1930s['year'].value_counts()
top_3_years = films_per_year.head(3)
print(top_3_years)
```

### Plot the number of films that have been released each decade over the history of cinema.

```python
import matplotlib.pyplot as plt

titles['decade'] = (titles['year'] // 10) * 10
films_per_decade = titles.groupby('decade').size()

plt.figure(figsize=(10, 6))
films_per_decade.plot(kind='bar', color='skyblue')
plt.title('Number of Films Released per Decade')
plt.xlabel('Decade')
plt.ylabel('Number of Films Released')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

### Plot the number of "Hamlet" films made each decade.

```python
hamlet_films = titles[titles['title'] == 'Hamlet']
hamlet_films['decade'] = (hamlet_films['year'] // 10) * 10
hamlet_per_decade = hamlet_films.groupby('decade').size()

plt.figure(figsize=(10, 6))
hamlet_per_decade.plot(kind='bar', color='lightgreen')
plt.title('Number of "Hamlet" Films Released per Decade')
plt.xlabel('Decade')
plt.ylabel('Number of "Hamlet" Films Released')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

### Plot the number of "Rustler" characters in each decade of the history of film.

```python
rustler_characters = cast[cast['character'] == 'Rustler']
rustler_characters['decade'] = (rustler_characters['year'] // 10) * 10
rustler_per_decade = rustler_characters.groupby('decade').size()

plt.figure(figsize=(10, 6))
rustler_per_decade.plot(kind='bar', color='lightblue')
plt.title('Number of "Rustler" Characters in Films per Decade')
plt.xlabel('Decade')
plt.ylabel('Number of "Rustler" Characters')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

### How many leading (n=1) roles were available to actors, and how many to actresses, in the 1950s?

```python
leading_roles_actors = cast[(cast['n'] == 1) & (cast['year'] // 10 == 195)]
leading_roles_actresses = cast[(cast['n'] == 1) & (cast['year'] // 10 == 195) & (cast['type'] == 'actress')]

num_leading_roles_actors = len(leading_roles_actors)
num_leading_roles_actresses = len(leading_roles_actresses)

print(f"Number of leading roles for actors in the 1950s: {num_leading_roles_actors}")
print(f"Number of leading roles for actresses in the 1950s: {num_leading_roles_actresses}")
```

### How many supporting (n=2) roles were available to actors, and how many to actresses, in the 1950s?

```python
supporting_roles_actors = cast[(cast['n'] == 2) & (cast['year'] // 10 == 195) & (cast['type'] == 'actor')]
supporting_roles_actresses = cast[(cast['n'] == 2) & (cast['year'] // 10 == 195) & (cast['type'] == 'actress')]

num_supporting_roles_actors = len(supporting_roles_actors)
num_supporting_roles_actresses = len(supporting_roles_actresses)

print(f"Number of supporting roles for actors in the 1950s: {num_supporting_roles_actors}")
print(f"Number of supporting roles for actresses in the 1950s: {num_supporting_roles_actresses}")
```
